# Exercício 01:

1. Execute a regressão em statsmodels para prever a variável 6. Para tal, use como exemplo o tutorial [daqui](https://www.statsmodels.org/stable/regression.html). Use apenas as variáveis age, sex, bmi e bp como preditoras.
1. Interprete os intervalos de confiança das variáveis. Existe alguma que poderia ser eliminada? Se sim, qual?

```python
# variáveis independentes e a variável dependente
X = df[["age", "sex", "bmi", "bp"]]
y = df["s6"]
X = sm.add_constant(X)

# modelo de regressão linear
model = sm.OLS(y, X).fit()
print(model.summary())
```

As variáveis age, sex, bmi e bp têm intervalos de confiança que não incluem zero. Portanto, nenhuma variável pode ser eliminada com base nos intervalos de confiança.

Já sobre valores p, "sex" é a variável com o maior valor p (0.013), mas ainda é menor que 0.05, então é significativa e não pode ser eliminada.

# Exercício 02:

1. Execute DUAS regressões para prever a quantidade de crimes violentos (violent) e quantidade de assassinatos (murder). Não use crimes violentos para prever assassinatos, nem assassinatos para precer crimes violentos. Use apenas as outras variáveis nos dois casos.
1. Interprete os intervalos de confiança das variáveis. Existe alguma que poderia ser eliminada? Se sim, qual?

```python
crime_data = sm.datasets.statecrime.load()
df = crime_data.data
df
```

```python
# regressão para prever a quantidade de crimes violentos
X_violent = df[["hs_grad", "poverty", "single", "white", "urban"]]
y_violent = df["violent"]
X_violent = sm.add_constant(X_violent)
model_violent = sm.OLS(y_violent, X_violent).fit()

print("Regressão para crimes violentos:")
print(model_violent.summary())
```

Todas as variáveis têm intervalos de confiança que incluem zero, exceto single. Além disso, single é a única variável com um p-valor significativamente menor que 0.05. Portanto, com base nos intervalos de confiança e nos valores p, poderíamos considerar eliminar as variáveis hs_grad, poverty, white e urban.

```python
# regressão para prever a quantidade de assassinatos
X_murder = df[["hs_grad", "poverty", "single", "white", "urban"]]
y_murder = df["murder"]
X_murder = sm.add_constant(X_murder)
model_murder = sm.OLS(y_murder, X_murder).fit()

print("\nRegressão para assassinatos:")
print(model_murder.summary())
```

Neste caso, as variáveis white e urban têm intervalos de confiança que incluem zero e p-valores maiores que 0.05, sugerindo que não são estatisticamente significativas para prever assassinatos. Portanto, com base nos intervalos de confiança e nos valores p, poderíamos considerar eliminar as variáveis white e urban do modelo.
